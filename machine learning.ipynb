```python
#subplot row wise
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,4,6,9,20])
y=np.array([3,6,7,10,35])
plt.subplot(211)
plt.plot(x,y)
plt.title("fi")
plt.show()
plt.subplot(212)
plt.plot(x,x*y)
plt.title("fixi")
plt.show()

```

```python
#4 subplots with grid color and plot color
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5])
plt.subplot(221)
plt.title("x=y")
plt.plot(x,x,"r")
plt.grid("g--")
plt.show()
plt.subplot(222)
plt.title("x=x^2")
plt.plot(x,x*x,"b")
plt.show()
plt.subplot(223)
plt.title("x=x^3")
plt.plot(x,x*x*x,"g")
plt.show()
plt.subplot(224)
plt.title("x=x^4")
plt.plot(x,x*x*x*x,"y")
plt.show()
```

```python
#colormap and color 
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5])
y= np.array([1,2,7,8,9])
colors=np.array([0,10,20,30,40])

plt.plot(x,y,linestyle="none")
plt.scatter(x,y,c=colors,cmap="viridis")
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([10,20,15,25,30])

sizes = np.array([.5,20,30,40,50])  # Bubble sizes

plt.scatter(x, y, s=sizes,color="r")

plt.title("Bubble Chart")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([10, 20, 15, 25])

plt.bar(
    x,
    y,
    color="skyblue",      # Bar color
    edgecolor="black",    # Border color
    linewidth=2,          # Border thickness
    width=0.6,            # Bar width
    alpha=0.8,            # Transparency
    hatch="//",           # Pattern
    label="Sales"         # Legend label
)

plt.title("Bar Chart Example")
plt.xlabel("Products")
plt.ylabel("Sales")

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.legend()

plt.show()
```
