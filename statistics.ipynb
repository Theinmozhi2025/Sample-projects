```python
#using statistics from python notebook itself
import statistics as stat
data=[12,56,68,96,24]
print("mean:",stat.mean(data))
print("median:",stat.median(data))
print("mode:",stat.mode(data))
print("variance:",stat.variance(data))
print("standard deviation:",stat.stdev(data))
print("population standard deviation",stat.pstdev(data))
```

```python
#using numpy(variance =var,stdev=std and no mode is used)
import numpy as np
data=[12,56,68,96,24]
print("mean:",np.mean(data))
print("median:",np.median(data))
#print("mode:",np.mode(data))
print("variance:",np.var(data))
print("standard deviation:",np.std(data))#default use of population standard deviation
```

```python
#using pandas
import pandas as pd
data=pd.Series([12,56,68,96,24])
print("mean:",data.mean())
print("median:",data.median())
print(data.mode())
print("variance:",data.var())
print("standard deviation:",data.std())#default in pandas(sample standard dev)
print("population standard deviation:",data.std(ddof=0))
```

```python
#using scipy(no median can be calculated so scipy uses numpy to calc)
from scipy import stats
data=[12,56,68,96,24]
print("mean",stats.tmean(data))
print("mode",stats.mode(data,keepdims=False))
print("Variance:", stats.tvar(data))
print("Standard Deviation:", stats.tstd(data))
```
